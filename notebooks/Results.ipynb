In the last notebooks. We tacled a classification problem for images. We tried to classify images as ship or not ship.

We managed to create a performant __Deep Learning__ model. We decided to use few metrics to measure the performance of the model.

Here are the important details that lead us to choose metrics from others :

* __Classification problem__
* __Class imbalance (75%, 25%)__

It was clear that having a such imbalance prevented us from using the accuracy as a metric.
The ship class as being the positive class that we want to predict, it lead us to focus on __AUROC__ and __Precision-Recall AUC__. Because they both focus on the positive class.

Precision-Recall AUC is an interesting chart because it allow us to gauge the tradeoff between the two important metrics. __As we focus on Recall__, it will allow us to select a compromise where the Recall is very high and the Precision as high as the model remains performant.

__This point needs to be specified with the experts, considering costs handling the errors.__
